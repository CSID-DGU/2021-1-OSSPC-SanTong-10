```python
import xml.etree.ElementTree as ET

tree = ET.parse('renjunet_v10_20180803.xml')
```

```python
root = tree.getroot()
```

```python
list_dim = root.findall('game/move')
list_dim
```

```python
for i in list_dim:
    print(i.text)
```

```python
def alpha2point(text):
    col = ord(text[0])-97
    raw = int(text[1:])-1
    return col, raw
```

# renju xml   ->   format 

```python
import numpy as np

no = 0
while(no < len(list_dim)):
    if no % 100 == 0:
        print(no)
    #else:
    #    print(no)
    
    inp = list_dim[no].text.split(' ')

    move_table = np.zeros([15, 15])

    num = 1
    for i in inp:
        if len(i) != 0:
            col, raw = alpha2point(i)
            move_table[raw, col] = num
            num += 1
        else: continue

    result = np.array(move_table).flatten()

    f = open('./data_YH/2021_04_30_{}.txt'.format(no+1), 'w')
    f.write(np.array2string(result))
    f.close()
    
    no += 1
```

```python
import os
file_path = 'training_data/txt/'
files = os.listdir(file_path)

remove_list = []
for file_name in files:
    file_name_path = os.path.join(file_path, file_name)
    
    board_vec = np.zeros([15*15], dtype='f')
    
    f = open(file_name_path, 'r')
    while True:
        line = f.readline()
        if not line: break
        split_ini = line.split('.')

        for i in range(len(split_ini)):
            split_fur = split_ini[i].split(' ')
            if split_fur[-1] != '\n':
                if split_fur[-1] == '[0':
                    remove_list.append(file_name_path)
    f.close()
for i in remove_list:
    os.remove(i)
```
