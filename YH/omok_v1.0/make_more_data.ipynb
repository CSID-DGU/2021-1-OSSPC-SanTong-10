```python
import pandas as pd
import numpy as np

file = open('renjunet_v10_20210509.txt', 'r')

game_data = []

for line in file:
    if line.startswith('<move>'):
        point = line[6:-8]
        game_data.append(point.split(' '))
game_data
```

```python
print(len(game_data))
```

```python
def alpha2point(text):
    col = ord(text[0])-97
    raw = int(text[1:])-1
    return col, raw
```

```python
tot_data_list = []

for pt in game_data:
    num = 1
    move_table = np.zeros([15, 15])
    for xy in pt:
        col, raw = alpha2point(xy)
        move_table[raw, col] = num
        num += 1
        
    result = np.array(move_table, dtype='int').flatten()
    
    tot_data_list.append(result)
```

```python
import pandas as pd

print(len(tot_data_list))

df = pd.DataFrame(tot_data_list)
df
```
