```python
import numpy as np

def data_15_15(list_data):
    list_data = np.array(list_data)
    new_form = np.reshape(list_data, (15,15))
    
    print(new_form)
    return new_form
```

```python
form = data_15_15([i for i in df.iloc[0]])
```

```python
def change_01(new_form):
    print(new_form)
    game_len = np.max(new_form)
    print(game_len)
    for idx1, row in enumerate(new_form):
        for idx2, col in enumerate(row):
            if col == game_len:
                if new_form[idx1, idx2] % 2 == 0:
                    new_form[idx1, idx2] = 4
                else:
                    new_form[idx1, idx2] = 3
            elif col == 0:
                new_form[idx1, idx2] = 0
            elif col % 2 == 0:
                new_form[idx1, idx2] = 2
            else:
                new_form[idx1, idx2] = 1
                
    print(new_form)
    return new_form
```

```python
new_form = change_01(form)
```

```python
def col_win(new_form):
    for col in range(15):
        col_list = new_form[:, col]
        if [1,1,1,1,1] in col_list:
            print('1')
        elif [2,2,2,2,2] in col_list:
            print('2')
        else:
            print('nothing')
```
