```python
import numpy as np

f = open('renjunet_v10_20210509.txt', 'r')

lines = f.readlines()

data_list = []
for line in lines:
    if line.startswith('<move>'):
        data_list.append(line[6:-8])

f.close()
data_list
```

```python
def alpha2point(text):
    col = ord(text[0])-97
    raw = int(text[1:])-1
    return col, raw
```

```python
for i in range(len(data_list)):
    inp = data_list[i].split(' ')
    move_table = np.zeros([15, 15])
    
    num = 1
    for j in inp:
        if len(j) != 0:
            col, raw = alpha2point(j)
            move_table[raw, col] = num
            num += 1
        else:
            continue
            
    result = np.array(move_table).flatten()
    
    f = open('./training_data/txt/v2_txt/2021_05_09_{}.txt'.format(i+1), 'w')
    f.write(np.array2string(result))
    f.close()
```
