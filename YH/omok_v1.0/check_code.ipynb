```python
import pandas as pd
import numpy as np

df = pd.read_csv('tot_data.csv')
```

```python
board_vec = np.zeros([15*15],dtype='f')

board_data = np.array(df.iloc[0][1:])
```

```python
board_stack = []

board_mat = board_data.reshape([15,15])

# rotate
board_stack.append(board_mat)
board_stack.append(np.rot90(board_mat))
board_stack.append(np.rot90(board_mat,2))
board_stack.append(np.rot90(board_mat,3))

board_fliplr = np.fliplr(board_mat)
board_flipud = np.flipud(board_mat)

# rotate and filp
board_stack.append(board_fliplr)
board_stack.append(np.rot90(board_fliplr))
board_stack.append(np.rot90(board_fliplr,2))
board_stack.append(np.rot90(board_fliplr,3))

board_stack.append(board_flipud)
board_stack.append(np.rot90(board_flipud))
board_stack.append(np.rot90(board_flipud,2))
board_stack.append(np.rot90(board_flipud,3))

board_stack
```

```python
board_size = 15
board_x_stack = []
board_y_stack = []


board_vec = board_stack[0].flatten() # 한 판의 결과를 (255,) 형태의 vector로
max_step = np.max(board_vec)
    
for data_cnt in range(1,int(max_step)): # 1~max_step
    board_x = np.zeros([board_size*board_size],dtype='f')
    board_y = np.zeros([board_size*board_size],dtype='f')
    
    for i in range(len(board_vec)): # 오목판의 각 좌표를 추출

        if data_cnt%2 != 0: #흑 player
            if board_vec[i] > 0 and board_vec[i] <= data_cnt: # 뽑아온 좌표가 n번째 착수 이전
                if board_vec[i]%2 != 0:
                    board_x[i] = 1 # 흑이면 X data에 1을 기입
                else:
                    board_x[i] = 2 # 백이면 X data에 2를 기입

            if board_vec[i] == data_cnt + 1: # 뽑아온 좌표가 n+1번째 착수일 경우
                board_y[i] = 1 # Y data에 1을 기입

        else:
            if board_vec[i] > 0 and board_vec[i] <= data_cnt:
                if board_vec[i]%2 == 0:
                    # 백
                    board_x[i] = 1
                else:
                    board_x[i] = 2

            if board_vec[i] == data_cnt + 1: 
                board_y[i] = 1     

    if (np.sum(board_y)) != 1:
        print('학습데이터에 문제가 있습니다.')

    board_x_stack.append(board_x/2)
    board_y_stack.append(board_y)
    
print(len(board_x_stack))
print(len(board_y_stack))
```

```python
board_x_stack
x_stack_li = []
for i in board_x_stack:
    x_stack = np.array(i).reshape(15,15)
    x_stack_li.append(x_stack)

for i in x_stack_li:
    print(i)
```

```python
board_y_stack
y_stack_li = []
for i in board_y_stack:
    y_stack = np.array(i).reshape(15,15)
    y_stack_li.append(y_stack)

for i in y_stack_li:
    print(i)
```
